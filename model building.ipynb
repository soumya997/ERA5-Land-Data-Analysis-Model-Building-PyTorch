# Model Building

```python
df2 = df.tail(50000)
```

```python
df3 = df2
df3.reset_index(inplace=True)
```

```python
df3.head()
```

```python
import datetime
df3['time'] = df3['time'].apply(lambda x: x.strftime('%Y-%m-%d %H:%M:%S'))
df3.head()
```

```python
df4 = df3.drop(['time'],axis=1)
df4.head()
```

```python
cols = ['latitude','longitude']
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

for i in cols:
    df4[i]=label_encoder.fit_transform(df4[i])
```

```python
df4.head()
```

```python
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
from sklearn.preprocessing import StandardScaler
import torch.nn.functional as F
from torch.autograd import Variable
```

```python
SEED = 42

data = df4.copy()
y = data['tp']
x = data.drop(['tp'],axis=1)


from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val = train_test_split(x,y,test_size = 0.2,random_state = SEED)
```

```python
type(x_train)
```

```python
x_train= np.array(x_train,dtype='float32')
y_train= np.array(y_train,dtype='float32')
x_val= np.array(x_val,dtype='float32')
y_val= np.array(y_val,dtype='float32')
```

```python
y_train2=y_train.reshape(40000,-1)
y_train2.shape
```

```python
sc_x = StandardScaler()
x_train_scaled = sc_x.fit_transform(x_train)
x_test_scaled = sc_x.transform(x_val)

sc_y = StandardScaler()
y_train_scaled = sc_y.fit_transform(y_train2)
# y_test_scaled = sc_y.transform(y_val)
```

```python
x_train_scaled
```

```python
x_train3 = torch.from_numpy(x_train_scaled).requires_grad_()
y_train3 = torch.from_numpy(y_train_scaled).requires_grad_()
```

```python
y_train3.shape
```

```python
batch_size = 64
train_ds = TensorDataset(x_train3, y_train3)
train_dl = DataLoader(train_ds, batch_size, shuffle=True)
```

```python
class LinearRegression(nn.Module):
    def __init__(self,input_size,hidden,output_size):
        # super function. It inherits from nn.Module and we can access everythink in nn.Module
        super(LinearRegression,self).__init__()
        # Linear function.
        self.linear1 = nn.Linear(input_dim,40)
        # self.relu1 = nn.ReLU()
        self.tanh1 = nn.Tanh()
        self.linear2 = nn.Linear(40,20)
        # self.relu2 = nn.ReLU()
        self.tanh2 = nn.Tanh()
        self.linear3 = nn.Linear(20,hidden)
        # self.drop = nn.Dropout(0.3)
        self.tanh3 = nn.Tanh()
        self.linear4 = nn.Linear(hidden,output_size)

    def forward(self,x):
      x = self.linear1(x)
      x = self.tanh1(x)
      x = self.linear2(x)
      x = self.tanh2(x)
      x = self.linear3(x)
      x = self.tanh3(x)
      x = self.linear4(x)
      # return x.squeeze()
      return x.squeeze()
    
# define model
input_dim = 4
hidden=10
output_dim = 1
our_model = LinearRegression(input_dim,hidden,output_dim) # 
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
device
```

```python
# our_model = SimpleNet() 
our_model.to(device)

# criterion = torch.nn.MSELoss(size_average = False) 
loss_fn = F.mse_loss
optimizer = torch.optim.Adam(our_model.parameters(), lr = 0.00001)
```

```python
y_train3.shape
```

```python
def fit(num_epochs, model, loss_fn, opt):
    for epoch in range(num_epochs):
      x_train1 = Variable(x_train3.cuda())
      y_train1 = Variable(y_train3.cuda())
      # Generate predictions
      pred = model(x_train1)
      loss = loss_fn(pred, y_train1)
      # Perform gradient descent
      loss.backward()
      opt.step()
      opt.zero_grad()
    # print('Training loss: ', loss_fn(model(x_train1), y_train1))
    if (epoch+1) % 10 == 0:

      print('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, num_epochs, loss.item()))
```
